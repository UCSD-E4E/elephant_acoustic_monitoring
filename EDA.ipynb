```python
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
import numpy as np
import librosa

def process_file(path):
    print(f"\n=== Processing {path} ===")

    # 1. Load the audio (same as your `y, sr = librosa.load(path=files[0])`)
    y, sr = librosa.load(path=path)

    # 2. Compute embeddings for 5-second (160000-sample) chunks
    embeds = []
    for i in range(0, y.shape[0], 160000):
        # same loop structure as in your notebook
        try:
            model_outputs = model.infer_tf(y[np.newaxis, i:i + 160000])
            embeds.append(model_outputs["embedding"])
        except Exception:
            # if a chunk fails, just skip it
            continue

    if not embeds:
        print("  No embeddings for this file (maybe too short), skipping.")
        return

    # 3. Turn list of embeddings into a 2D array, same as:
    #    X = np.array(embeds); X = X[:, 0, :]
    X = np.array(embeds)
    if X.ndim == 3:          # what your original code expects
        X = X[:, 0, :]

    if X.shape[0] <= 3:
        print("  Not enough chunks for t-SNE (<=3), skipping.")
        return

    # 4. t-SNE (same hyperparameters as your notebook)
    X_embedded = TSNE(
        n_components=2,
        learning_rate="auto",
        init="random",
        perplexity=3,
    ).fit_transform(X)

    # 5. DBSCAN clustering (same as your notebook)
    cluster = DBSCAN(min_samples=2, eps=4)
    labels = cluster.fit_predict(X_embedded)

    # 6. Print a quick summary so you know it ran
    unique, counts = np.unique(labels, return_counts=True)
    print("  #chunks:", X.shape[0])
    print("  clusters:", dict(zip(unique, counts)))
```

```python
for f in test_files:
    process_file(f)
```

```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

CHUNK_SIZE = 160000   # 5 seconds at 32 kHz

# Pick a cluster label you want to visualize (change 17 → any cluster)
target_label = 17

# Get all chunk indices belonging to that cluster
chunk_indices = np.where(labels == target_label)[0]

# Safety check
if len(chunk_indices) == 0:
    print(f"No chunks found for cluster {target_label}")
else:
    # Pick one random chunk index
    chosen_chunk = np.random.choice(chunk_indices)

    # Convert chunk index → sample start index
    start = chosen_chunk * CHUNK_SIZE

    # Extract exactly one 5-second chunk
    snippet = y[start:start + CHUNK_SIZE]

    # Compute STFT
    D = librosa.stft(snippet)
    S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

    # Plot
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(S_db, sr=sr, x_axis="time", y_axis="hz")
    plt.title(f"Cluster {target_label} — Chunk {chosen_chunk}")
    plt.colorbar(label="dB")
    plt.tight_layout()
    plt.show()
```
