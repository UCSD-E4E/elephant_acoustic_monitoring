```python
from matplotlib import pyplot as plt
import numpy as np
import tensorflow_hub as hub
import tensorflow as tf
tf.experimental.numpy.experimental_enable_numpy_behavior()

# Load the model.
model = hub.load('https://www.kaggle.com/models/google/bird-vocalization-classifier/TensorFlow2/bird-vocalization-classifier/8')
```

```python
embeds = []
for i in range(0, y.shape[0], 160000):
    print(i)
    try:
        model_outputs = model.infer_tf(y[np.newaxis, i:i + 160000])
        embeds.append(model_outputs['embedding'])
    except Exception as e:
        continue
```

```python
from sklearn.manifold import TSNE

X = np.array(embeds)
X = X[:, 0, :]

X_embedded = TSNE(n_components=2, learning_rate='auto',

                  init='random', perplexity=3).fit_transform(X)

X_embedded.shape
```

```python
plt.scatter(X_embedded[:, 0], X_embedded[:, 1])
```

```python
from sklearn.cluster import DBSCAN
```

```python
cluster = DBSCAN(min_samples=2, eps=4)
labels = cluster.fit_predict(X_embedded)
```

```python
labels
```

```python
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=labels)
plt.colorbar()
```
